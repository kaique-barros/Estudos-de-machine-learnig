```python
import numpy as np
import pandas as pd
```

Após a etapa de tratamento dos dados, agora vamos começar o pré-processamento deles para que os algoritmos de machine learning possa interpreta-los corretamente.

```python
dados = pd.read_csv('heart_tratado.csv', sep=';', encoding='utf-8')
dados.head()
```

```python
dados.describe()
```

```python
dados.shape
```

## Transformando as variáveis categóricas nominais em variáveis categóricas oridinais

```python
df2 = dados.copy()
df2['Sex'] = df2['Sex'].map({'M': 0, 'F': 1})
df2['ChestPainType'] = df2['ChestPainType'].map({'ATA': 0, 'NAP':1, 'ASY':2, 'TA':3})
df2['RestingECG'] = df2['RestingECG'].map({'Normal': 0, 'ST':1, 'LVH':2})
df2['ExerciseAngina'] = df2['ExerciseAngina'].map({'N': 0, 'Y':1})
df2['ST_Slope'] = df2['ST_Slope'].map({'Up': 0, 'Flat':1, 'Down':2})
```

```python
df2
```

## Atributos previsores e alvos

O objetivo do algorimo é prever quais pessoas tem uma maior tendencia a ter problemas cardiacos, então o alvo é HeatDesease e todos os outros parametros são previsores

```python
previsores = df2.iloc[:, 0:11].values
previsores
```

```python
previsores.shape
```

```python
alvo = df2.iloc[:, 11].values
alvo
```

## Codificação das variáveis categóricas

```python
from sklearn.preprocessing import LabelEncoder
```
