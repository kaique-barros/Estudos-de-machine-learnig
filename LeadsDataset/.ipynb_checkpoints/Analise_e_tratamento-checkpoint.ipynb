## Análise das Colunas

### Prospect ID e Lead Number

Ambas as colunas servem apenas para organização dos dados e tem pouco valor quando falamos da conversão de um potencial cliente, por isso inicialmente vou deixa-las de lado

### Lead Origin

```python
df['Lead Origin'].value_counts().reset_index()
```

Lead Origin é uma variável qualitativa ordinal e indica de onde o Lead surgiu. 

É interessante observar que existem:
- Dois valores que agrupam a maior parte dos dados: 'Landing Page Submission' e 'Api'
- Dois valores com menos entradas mas ainda com dados suficientes para ser considerados relevantes: 'Lead Add Form' e 'Lead Import'
- E um valor que possui apenas uma entrada: 'Quick Add Form'

Manter esse valor único pode causar um viés no modelo, o melhor maneira de lidar com ele é agrupa-lo com 'Lead Add Form', isso porque seus contextos são muito semelhantes.

```python
# Agrupando 'Quick Add Form' com 'Lead Add Form'
df['Lead Origin'] = df['Lead Origin'].replace('Quick Add Form', 'Lead Add Form')
```

```python
df['Lead Origin'].value_counts().reset_index()
```

```python
LO_graph = (sns.histplot(df, x='Lead Origin', hue='Lead Origin', bins=30, stat='count')
                        .set_xticklabels([])
)
```

### Lead Source

```python
df['Lead Source'].value_counts().reset_index()
```

Lead Source representa onde o visitante concheceu o produto.  
Ela é um caso levemente diferente de Lead Origin: apesar de ambos representarem variáveis qualitativas ordinais a **source tem muitos casos com poucos dados para serem considerados relevantes**, então **apesar individualmente representarem poucos dados, quando agrupados representam uma parte importante do dataset**.  
Por isso quando tratamos eles vamos buscar formas de agupar dados com poucas ocorrencias em categorias maiores, por exemplo:  
- 'bing' e 'google' podem criar uma categoria de 'buscadores'.  
- 'Facebook' pode ser agrupado em 'Social Media'.

Porém nesse caso, mesmo que os agrupemos em categorias dessa forma a quantidade de dados em cada categoria ainda seria pouca, o que pode acarretar em overfitting.  
Por isso a solução é **definir um limite**, em que **qualquer categoria com menos dados que esse limite será agrupada em uma categoria 'Others'**. Um bom limite para o nosso caso é a categoria 'Referral Sites', pois comparativamente é a última categoria com um número relevante de dados.

```python
contagem_LS = df['Lead Source'].value_counts()
limite = df['Lead Source'].value_counts().get('Referral Sites', 0)

categorias_para_troca = contagem_LS[contagem_LS < limite].index

df.loc[df['Lead Source'].isin(categorias_para_troca), 'Lead Source'] = 'Others'
df['Lead Source'].value_counts().reset_index()
```

```python
LS_graph = (sns.histplot(df, x='Lead Source', hue='Lead Source', bins=30, stat='count')
                        .set_xticklabels([])
)
```

### Total Visits

```python
df['TotalVisits'].value_counts().reset_index()
```

'TotalVisits' é a nossa primeira variável quantitativa, e por isso necessita de uma análise um pouco mais aprofundada. Primeiramente, vamos analisar a média e a amplitude:

```python
print(f'Média: {df['TotalVisits'].mean():.2f}')
print(f'Amplitude: {max(df['TotalVisits']) - min(df['TotalVisits'])}')
```

Inicialmente podemos notar uma grande amplitude, o que não seria algo importante contanto que os dados estivessem bem distribuidos, porém note a média: a média dos dados esta próxima a 3 porém a amplitude é 251, o que significa que temos um ou alguns outilers muito distantes na distribuição.  
Vamos observar as medidas de dispersão para analisar como os outliers estarão distribuídos:

```python
print(f'Variância: {df['TotalVisits'].var(ddof=1):.2f}')
print(f'Desvio Padrão: {df['TotalVisits'].std():.2f}')
print(f'Coeficiente de Variação: {df['TotalVisits'].std() / df['TotalVisits'].mean() * 100:.2f}%')
```

Como visto, o **desvio padrão esta muito alto** (considerando a média de 3.45), isso é ainda mais acentuado quando olhamos o coeficiente de variação de 141%.  
Podemos esperar que a maior parte dos valores estejam entre 0 - 8 e alguns outliers muito distantes da média o que esta causando a amplitude tão distante da média.  
Vamos observar os gráficos de usuários por número de acessos, o que esperamos é um pico no começo e uma rápida queda após ele:  

```python
#sns.histplot(df[df['TotalVisits'] > 0], x='TotalVisits', kde=True, log_scale=True)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico 1: Visão Geral
sns.histplot(df['TotalVisits'], kde=True, ax=ax1)
ax1.set_title('Visão Geral (0-250)')
#ax1.axvline(df['TotalVisits'].mean(), color='red', label='Média')

# Gráfico 2: Visão Detalhada
sns.histplot(df[df['TotalVisits'] <= 20]['TotalVisits'], kde=True, ax=ax2, color='orange')
ax2.set_title('Detalhe do Pico (0-20)')
#ax2.axvline(df['TotalVisits'].mean(), color='red', label='Média')

plt.tight_layout()
plt.show()
```

O gráfico está exatamente como o previsto, um grande pico no começo, uma caída abrupta após e uma longa cauda. Mais uma vez precisamos tratar esses outliers para evitar bias, no caso de uma variavel discreta como essa temos 2 opções de tratamento:  
- Substituir os outliers pela média, ou;
- Substituir os outilers pelo último valor antes de um limite definido.

Para decidir qual opção devemos escolher, primeiro vamos ter uma ideia da quantidade de outliers:

```python
px.box(df, 'TotalVisits')
```

O gráfico de caixa nos mostra os quartis do conjunto de dados, assim como os limites superior e inferior.  

Os outilers são os valores após a barra direita do gráfico, podemos perceber que a maioria deles ainda está bem próxima do limite superior (11), e, apesar de ainda ser muito extremo temos uma certa progressão até chegar ao maior valor do conjunto de dados (251).  
Isso me leva a pensar que, **apesar de discrepante, o dado não é um erro de medição**, por isso ainda devemos manter sua característica, mas não podemos manter o mesmo valor para evitar overfitting.  

Por isso, primeiramente **vou testar a técnica de capping** (achatamento), ou seja, vou definir um valor do quantil 99% e todos os valores acima dele serão achatados para esse valor.

Mantenha em mente que **essa estratégia ainda pode causar distorção no modelo**, e, caso ache que é absolutamente necessário manter a hierarquia dos dados a transformação por escala logarítimica é melhor.

```python
limite = df['TotalVisits'].quantile(0.99)
df['TotalVisits_Capped'] = df['TotalVisits'].clip(upper=limite)
df['TotalVisits_Capped'].value_counts().reset_index()
```

```python
print(f'Média: {df['TotalVisits_Capped'].mean():.2f}')
print(f'Amplitude: {max(df['TotalVisits_Capped']) - min(df['TotalVisits_Capped'])}')
print(f'Variância: {df['TotalVisits_Capped'].var(ddof=1):.2f}')
print(f'Desvio Padrão: {df['TotalVisits_Capped'].std():.2f}')
print(f'Coeficiente de Variação: {df['TotalVisits_Capped'].std() / df['TotalVisits_Capped'].mean() * 100:.2f}%')
```

Após o capping podemos perceber que o coeficiente de variação teve uma grande queda, porém ainda é considerado alto. Isso sugere uma estratégia de segmentação personalizada, ou seja, segmentar os dados em grupos como:
- Curiosos (de 1 - 5 visitas)
- Interessados ( de 6 - 11)
- Muito interessados ( 12+ )  

E então tratar cada um dos grupos como individualmente com estratégias e investimento diferente a cada um dos grupos, entretanto

```python
sns.histplot(df['TotalVisits_Capped'], kde=True)
```
